# 1. List Operations and Basic Statistics

```python
l = [i for i in range(1, 21)]

sumEven = 0
for i in l:
    if i % 2 == 0:
        sumEven += i
print("Sum of even numbers:", sumEven)

prodOdd = 1
for i in l:
    if i % 2 != 0:
        prodOdd *= i
print("Product of odd numbers:", prodOdd)

mean = sum(l)/20
median = l[10]

diff = 0
for i in l:
    diff += (i-mean)**2

SD = (diff/20)**0.5

print(f"Mean: {mean}, Median: {median}, StandardDeviaition: {SD}")
```

# 2. Mutable vs. Immutable Data Types

Mutable datatypes can be edited even after creating them and can make changes to the values of it eg: Lists, Dictionaries.

Immutable on the other hand is fixed and no change in values can be made after creation. eg: strings, tuples

# 3. Fibonacci Sequence

```python
import matplotlib.pyplot as plt
```

```python
def fibonaci(n: int):
    l = [0, 1]
    for i in range(n-2):
        l.append(l[-1]+l[-2])
    return l

plt.plot(fibonaci(20))
```

# 7. Dictionary Operations

```python
#students
marks = {"Alpha": 76, "Beta": 83, "Charlie": 92}

avg = sum(marks.values())/len(marks)
print(f"average: {avg}")

for i in marks:
    if marks[i] > avg:
        print(i)

plt.bar(marks.keys(), marks.values())
plt.show()
```

# 9. Merging Dictionaries

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}

out = dict1
for i in dict2:
    out[i] = dict1.get(i, 0) + dict2[i]
print(out)
```

# 10. Character Frequency

# 11. Remove Duplicates

```python
import statistics as stat
```

```python
original = [1,4,5,6,4,3,7,8,9,5,6,1,3]
unique = []

for i in original:
    if i not in unique:
        unique.append(i)
print(unique)
print(f"mean: {stat.mean(unique)}, median: {stat.median(unique)}")
```
